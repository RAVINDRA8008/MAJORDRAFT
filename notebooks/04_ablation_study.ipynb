```python
import sys, os
os.chdir('/content/amers')
sys.path.insert(0, '.')

from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np

DRIVE_BASE = Path('/content/drive/MyDrive/AMERS')
OUT = DRIVE_BASE / 'outputs'
```

```python
# Ablation results (fill in after running ablation experiments)
# You can also load from results.json if saved
import json

json_path = OUT / 'results.json'
if json_path.exists():
    with open(json_path) as f:
        all_results = json.load(f)
    ablation = all_results.get('ablation', {})
else:
    print('No results.json found. Run evaluate.py first.')
    ablation = {}

if ablation:
    names = list(ablation.keys())
    accs = [ablation[n].get('accuracy', 0) for n in names]
    f1s = [ablation[n].get('f1_macro', 0) for n in names]
    
    x = np.arange(len(names))
    width = 0.35
    
    fig, ax = plt.subplots(figsize=(12, 5))
    ax.bar(x - width/2, accs, width, label='Accuracy')
    ax.bar(x + width/2, f1s, width, label='F1 (macro)')
    ax.set_xticks(x)
    ax.set_xticklabels(names, rotation=45, ha='right')
    ax.set_ylabel('Score')
    ax.set_title('Ablation Study Results')
    ax.legend()
    ax.set_ylim(0, 1)
    plt.tight_layout()
    plt.savefig(str(OUT / 'ablation_chart.png'), dpi=150)
    plt.show()
```
